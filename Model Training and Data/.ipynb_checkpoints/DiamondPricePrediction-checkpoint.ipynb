# Model Building

```python
X = labelData.drop(["price"],axis=1)
y = labelData[("price")]
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25, random_state=7)
```

```python
from sklearn import set_config
```

```python
set_config(display="diagram")
```

```python
pipelineLR = Pipeline([("scalar1", StandardScaler()), ("LRClassifier",LinearRegression())])
pipelineDT = Pipeline([("scalar2", StandardScaler()), ("DTClassifier",DecisionTreeRegressor())])
pipelineRF = Pipeline([("scalar3", StandardScaler()), ("RFClassifier",RandomForestRegressor())])
pipelineKN = Pipeline([("scalar4", StandardScaler()), ("KNClassifier",KNeighborsRegressor())])
pipelineXBG = Pipeline([("scalar5", StandardScaler()), ("XBGClassifier",XGBRegressor())])

# pipelines = [pipelineLR, pipelineDT, pipelineRF, pipelineKN, pipelineXBG]

pipelines = [pipelineLR, pipelineDT, pipelineRF, pipelineKN, pipelineXBG]
pipeline_names = ["Linear Regression", "Decision Tree", "Random Forest", "K-Neighbors", "XGBoost"]

# Visualizing all pipelines
for name, pipe in zip(pipeline_names, pipelines):
    print(f"Pipeline: {name}")
    display(pipe)  # Visualizes each pipeline
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

results = []

for name, pipe in zip(pipeline_names, pipelines):
    pipe.fit(X_train, y_train)
    y_train_pred = pipe.predict(X_train)
    r2 = r2_score(y_train, y_train_pred)
    rmse = np.sqrt(mean_squared_error(y_train,y_train_pred))
    results.append({"Model": name, "R^2 Score": r2, "RMSE": rmse})

    print(f"Model: {name}")
    print(f"R^2 score: {r2: .4f}")
    print(f"RMSE: {rmse: .4f}")
    print("-" * 30)
    
```

```python

results = []

for name, pipe in zip(pipeline_names, pipelines):
    pipe.fit(X_train, y_train)
    y_test_pred = pipe.predict(X_test)
    r2 = r2_score(y_test, y_test_pred)
    rmse = np.sqrt(mean_squared_error(y_test,y_test_pred))
    results.append({"Model": name, "R^2 Score": r2, "RMSE": rmse})

    print(f"Model: {name}")
    print(f"R^2 score: {r2: .4f}")
    print(f"RMSE: {rmse: .4f}")
    print("-" * 30)
```
